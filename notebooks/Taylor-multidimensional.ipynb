# Taylorutvikling i 2 dimensjoner

Vi ser på funksjonen

$$
f(x, y) = (1-y^2)\cos(x) 
$$

og finner at Taylorutviklingen til andre orden er

$$
T(x, y) = 1-\frac{1}{2}x^2-y^2
$$

Under plotter vi både $f$ og $T$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline
```

```python
t = np.linspace(-5, 5, 21)
x, y = np.meshgrid(t, t, indexing="ij")
f = (1-y**2)*np.cos(x)
T = 1-0.5*x**2-y**2
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
surf = ax.plot_surface(x, y, f, cmap=plt.cm.jet)
surf = ax.plot_surface(x, y, T)
```
