# Strømlinjer

Vi har gitt et vektorfelt

$$
\vec{v} = -y \mathbf{i} + x \mathbf{j}
$$

Finn strømlinjene!

Løser $\vec{v} \times d\vec{r}=0$ hvilket gir

$$
x^2 + y^2 = C
$$

der $C$ er en konstant.

Vi kan plotte $\vec{v}$ med [quiver](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.quiver.html) og strømlinjene med [contour](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.contour.html)

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
t = np.linspace(-5, 5, 10)
x, y = np.meshgrid(t, t, indexing="ij")
plt.figure(figsize=(5, 5))
plt.quiver(x, y, -y, x, pivot='middle')
plt.contour(x, y, x**2+y**2, 8)
```

Plott med finere sirkler, men like mange vektorer. Bruk 4 ganger så mange punkter i hver retning, men velg kun hvert fjerde punkt til vektorpilene. 

```python
t = np.linspace(-5, 5, 40)
x, y = np.meshgrid(t, t, indexing="ij")
plt.figure(figsize=(5, 5))
plt.quiver(x[::4, ::4], y[::4, ::4], -y[::4, ::4], x[::4, ::4], pivot='middle')
plt.contour(x, y, x**2+y**2, 8)
```
