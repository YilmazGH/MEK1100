# Terreng

Et terreng er gitt ved

$$
h(x,y) = x^2-y^2
$$

der $x$ peker mot øst og $y$ mot nord. Finn stigningstallet til $h$ i punktet $(1, 0)$ i retning nordøst.

Retning nordøst er gitt ved vektoren $\vec{u} = (\mathbf{i} + \mathbf{j})/\sqrt{2}$.

Starter med å plotte terrenget i både et 3D plot og et kontourplot.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline
t = np.linspace(-5, 5, 21)
x, y = np.meshgrid(t, t, indexing="ij")
h = x**2-y**2
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
surf = ax.plot_surface(x, y, h, cmap=plt.cm.jet)
```

```python
fig2 = plt.figure()
plt.contour(x, y, h, 10)
```

Finner stigningstallet ved punktet (1, 0) gjennom å evaluere

$$
\frac{dh}{ds} = \nabla h \cdot \vec{u}
$$

ved $(1, 0)$. Finner først $\nabla h$ og definerer $\vec{u}$

```python
import sympy as sp
from sympy.vector import CoordSys3D
x, y = sp.symbols('x,y')
h = x**2-y**2
N = CoordSys3D('N')
gradh = h.diff(x, 1)*N.i + h.diff(y, 1)*N.j
u = (N.i+N.j)/np.sqrt(2)
dhds = gradh.dot(u)
```

Evaluerer ved $(1, 0)$. Stigningstallet er

```python
dhds.subs({'x':1, 'y':0})
```
