# Taylorutvikling i én dimensjon

Gitt funksjon

$$
f(x) = \cos(x)
$$

Finn Taylor-utviklingen til vilkårlig orden rundt $x=0$.

```python
import sympy as sp
x = sp.symbols('x')
f = sp.cos(x)
```

```python
sp.series(f, x, x0=0, n=12)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

xj = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.figure()
plt.plot(xj, np.cos(xj), 'b', linewidth=3)
for n in range(2, 12, 2):
    T = sp.series(f, x, x0=0, n=n).removeO()
    plt.plot(xj, [T.subs({'x': h}) for h in xj])
plt.axis([-2*np.pi, 2*np.pi, -2, 2])
plt.legend(['cos(x)']+list(range(2, 12, 2)), loc='upper left')
h = plt.title('Taylorutvikling av cos(x)')
```
